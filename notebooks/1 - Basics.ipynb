```python
from clip_saver import ClipSaver
from clip_saver.video_source import FileVideoSource
from clip_saver.callbacks import PrintInfoCallback, VideoSaverCallback
```

```python
saver = ClipSaver(
    model_path="yolo12n.pt",
    video_source=FileVideoSource(
        file_path="https://www.youtube.com/watch?v=EHTah-w6ihw"
    ),
    callbacks=[
        PrintInfoCallback(),
        VideoSaverCallback(output_path="test.mp4"),
    ],
    model_kwargs={
        "conf": 0.25,
        "show": True,
        "device": "mps",
    },
)

saver.start()
```

```python
saver.stop()
```

NOTE: we keyboard interupt to avoid creating a large test.mp4 output.
